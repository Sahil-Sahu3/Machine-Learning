## 2. The question: decision, action, cost of a wrong call

The Decision: Figuring out which of our top-performing web pages are losing traffic or dropping on Google, so we can put them at the top of our to-do list to fix.

The Action: Our writing and SEO team will take the worst-performing pages from the top of this list, look at the notes explaining why they are struggling, and jump in to update the info, add more details, or fix the keywords.

The Cost of a Wrong Call:

If we flag a healthy page by mistake (False Positive): We waste a writer's valuable time editing an article that was already doing perfectly fine. That is money down the drain for zero reward.

If we miss a dying page (False Negative): We let a highly profitable page silently crash down Google's search results. Once a competitor takes that top spot from us, it is incredibly hard, and expensive, to win those readers back.
